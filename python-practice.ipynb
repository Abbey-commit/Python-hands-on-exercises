Exercise 1: Create a Python program that takes two numbers as inputs from
 the user, performs all basic arithmetic operations on these numbers, and
 prints the results.

```python
# taking two numbers as input from users
num1 = int(input("Input the first number: "))
num2 = int(input("Input the second number: "))

# perform the operation on each of them
subtract = num1 - num2
add = num1 + num2
divide = num1 / num2
multiply = num1 * num2
modulo = num1 % num2
exponent = num1 ** num2
floor_divide = num1 // num2

# print the result
print(f"The subtraction of second number {num2} from the first number {num1} is: {subtract}.")
print(f"The sum of {num1} and {num2} is: {add}")
print(f"The division of {num1} by {num2} s: {divide}.")
print(f"The multiplication of {num1} by {num2} is: {multiply}.")
print(f"The remainder in the division of {num1}, {num2}, is: {modulo}.")
print(f"The exponential of {num1} and {num2} is: {exponent}.")
print(f"The floor division of {num1} by {num2} is: {floor_divide}.")
```

```python
class MyIterator:
    def __init__(self):
        self.a = 1
        
    def __iter__(self):
        return self
        
    def __next__(self):
        x = self.a
        self.a += 1
        return x

# Create an object of the iterator
my_iter = MyIterator()

# Use next() to get the next items in the iterator
print(next(my_iter)) # Output: 1
print(next(my_iter)) # Output: 2
```

```python
# Generator
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

count = count_up_to(5)
count
```

```python
print(next(count))
print(next(count))
print(next(count))
print(next(count))
print(next(count))
```

```python
counter = count_up_to(9)

for item in counter:
    print(item)
```

Iterables and Iterators
 Create a Python iterator that returns the Fibonacci series. The Fibonacci
 sequence is a series of numbers in which each number is the sum of the two
 preceding ones, usually starting with 0 and 1.

```python
# Here's a sample solution
class Fibonacci:
    def __init__(self):
        self.a = 0
        self.b = 1
    def __iter__(self):
        return self

    def __next__(self):
        fib = self.a
        self.a, self.b = self.b, self.a + self.b
        return fib

fib = Fibonacci()
for i in range(10):
    print(next(fib), end=' ')
```
